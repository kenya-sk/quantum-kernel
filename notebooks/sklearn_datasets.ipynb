## Visualize and Basic Analysis for `scikit-learn` Classification Datasets

Official Document: https://scikit-learn.org/stable/api/sklearn.datasets.html

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn import datasets
```

```python
RONDOM_STATE = 42

DATA_HOME_PATH = "../data/"
# datasets.clear_data_home(DATA_HOME_PATH)  
```

#### Dataset 1: fetch_20newsgroups

```python
news_group_dataset = datasets.fetch_20newsgroups(subset='train', categories=None, random_state=RONDOM_STATE, data_home=DATA_HOME_PATH)
print(news_group_dataset.DESCR)
```

```python
print(f"Train Data Size: {len(news_group_dataset.data)}")
print(f"Unique Target Classes: {len(news_group_dataset.target_names)}")
print(f"Target Classes: {news_group_dataset.target_names}")
print("-----------------------------------------------")
print(f"Data Sample: {news_group_dataset.data[0]}")
print(f"Target Class: {news_group_dataset.target_names[news_group_dataset.target[0]]}")
```

#### Dataset 2: fetch_20newsgroups_vectorized

Vectorized dataset of text data contained in fetch_20newsgroups. Various data and usage are the same.

#### Dataset 3: fetch_covtype

```python
covtype_dataset = datasets.fetch_covtype(data_home=DATA_HOME_PATH, random_state=RONDOM_STATE, as_frame=True)
print(covtype_dataset.DESCR)
```

```python
covtype_df = covtype_dataset.frame
print(f"Dataset Shape: {covtype_df.shape}")
print(f"Unique Target Classes: {covtype_df['Cover_Type'].nunique()}")

covtype_df.head()
```

```python
print(covtype_df.describe())
```

```python
cols = covtype_dataset.feature_names[:5]
target_col = covtype_dataset.target_names[0]
target_class = [1, 2, 3, 4, 5]
df = covtype_df[covtype_df[target_col].isin(target_class)][cols + [target_col]].sample(1000)
sns.pairplot(df, hue=target_col)
plt.show()
```

#### Dataset 4: fetch_kddcup99

original dataset is multi-class classification. 

but, we can convert it to binary classification by formattting the target column as attack or normal

```python
# kdd_dataset = datasets.fetch_kddcup99(subset='SA', data_home=DATA_HOME_PATH, random_state=RONDOM_STATE, percent10=True, as_frame=True)
kdd_dataset = datasets.fetch_kddcup99(subset='http', data_home=DATA_HOME_PATH, random_state=RONDOM_STATE, percent10=True, as_frame=True)
print(kdd_dataset.DESCR)
```

```python
kdd_df = kdd_dataset.frame
print(f"Dataset Shape: {kdd_df.shape}")
print(f"Unique Target Classes: {kdd_df['labels'].nunique()}")
print(f"Target Classes: {kdd_df['labels'].unique()}")

kdd_df.head()
```

```python
print(kdd_df.describe())
```

```python
df = kdd_df.sample(50000)
df['labels'] = df['labels'].apply(lambda x: 'normal' if x == b'normal.' else 'attack')

sns.pairplot(df, hue="labels")
plt.show()
```

#### Dataset 6: load_digits

```python
digits_dataset = datasets.load_digits()
print(digits_dataset.DESCR)
```

```python
sample_num = 10

plt.figure(figsize=(10, 5))
for i in range(sample_num):
    plt.subplot(1, sample_num, i + 1)
    plt.imshow(digits_dataset.images[i], cmap="gray")
    plt.axis('off')
```
